### Cleaning up the dataframe:

#### Duplicated Rows:

```python
# checking the number of duplicated rows:

df.duplicated().sum()
```

```python
#dropping the duplicated rows and storing the resulting entities in a new dataframe:

clean_df = df.drop_duplicates()
```

We can see that the apps differ mostly by the number of reviews and categories. I will drop the duplicates according to the number of reviews. 

We will be keeping the entry with the highest number of reviews!

To do so we have to go through a number of steps:

```python
# sorting the df by app name and number of reviews:

clean_df = clean_df.sort_values(by=['App', 'Reviews'], ascending=[True, False])
```

```python
# dropping duplicates on the App name level and just keeping the first entry which will be the one with the highest number of reviews:

clean_df = clean_df.drop_duplicates(subset=['App'], keep='first')
```

```python
# making sure this worked:

clean_df['App'].duplicated().sum()
```

now that we are done with duplicates, let us see what we can do with the null values:

### Fixing Up Columns + Dtypes:

we can fill this by external research!

With research, we found out that this game is free!

```python
# having a general look at the data again:

clean_df.head()
```

#### Category Column:

```python
# counting the number of unique categories:

clean_df['Category'].nunique()
```

```python
# getting the names of the categories:

clean_df['Category'].value_counts()
```

```python
# looking at the unique categories names:

clean_df['Category'].unique()
```

we can see in some categories that the space is replaced with a '_', let us fix that:

```python
# replacing '_' with ' ':

clean_df['Category'] = clean_df['Category'].str.replace('_', ' ')
```

```python
# checking if it worked:

clean_df['Category'].value_counts()
```

#### Size Column:

```python
# looking at the values in the Size column:

clean_df['Size'].value_counts()
```

We decided to turn the one with mb in size to kb, and turn varies with device to 0!

```python
# cleaning up the data in the size column:

clean_df['Size'] = clean_df['Size'].str.replace('Varies with device', '0')
clean_df['Size'] = clean_df['Size'].str.replace('M', '000')
clean_df['Size'] = clean_df['Size'].str.replace('.', '')
clean_df['Size'] = clean_df['Size'].str.replace('k', '')
```

```python
# turning the data type from string to integer: 

clean_df['Size'] = clean_df['Size'].astype(int)
```

#### Reviews Column:

we just have to change the data type of the reviews column!

```python
# changing the data type of the reviews column into an integer:

clean_df['Reviews'] = clean_df['Reviews'].astype(int)
```

#### Price Column:

We fix it up first removing the $ sign ang then tunring the price into a numeric

```python
# removing the $ sign

clean_df['Price'] = clean_df['Price'].str.replace('$', '')
```

```python
# changing the price column dtype to float:

clean_df['Price'] = clean_df['Price'].astype(float)
```

#### Last Updated Column:

```python
# changing the data type of the last updated column to datetime:

clean_df['Last Updated'] = pd.to_datetime(clean_df['Last Updated'], format='mixed')
```

#### Removing Some Columns:

we will be dropping the android ver and current ver columns as they will not be used for the ML model or for analysis even!

```python
# dropping the android ver and current ver columns:

clean_df = clean_df.drop(['Android Ver', 'Current Ver'], axis=1)
```

#### Installs Column:

we will be turning the installs column into the label column

```python
# looking at the unique values of the installs column:

clean_df['Installs'].value_counts() 
```

```python
# cleaning up the installs column:

clean_df['Installs'] = clean_df['Installs'].str.replace('+', '')
clean_df['Installs'] = clean_df['Installs'].str.replace(',', '')
```

```python
# changing the data type from string to integer:

clean_df['Installs'] = clean_df['Installs'].astype(int)
```

 - from 0 to 10,000: not successful -> 0
 - from 10,000 to 100,000:  slightly successful -> 1
 - from + 100,000: successful -> 2

```python
clean_df['Installs'].loc[clean_df['Installs'] < 10000] = 0
clean_df['Installs'].loc[(clean_df['Installs'] >= 10000) & (clean_df['Installs']<100000)] = 1
clean_df['Installs'].loc[(clean_df['Installs'] >= 100000)] = 2
```

```python
# checking if the above step worked!

clean_df['Installs'].value_counts()
```

the data is not 50-50 but is close enough to 50-50 which is good and still abit realistic!

```python
# renaming the installs column to label:

clean_df.rename(columns={'Installs': 'Label'}, inplace=True)
```

#### Type Column:

we will be changing the type column to boolean like where if the app is free the rows entry will be 0, if it is paid the rows entry will be 1

```python
# mapping the entries in the column: free to 0, paid to 1:

clean_df['Type'] = clean_df['Type'].map({'Free': 0, 'Paid': 1})
```

```python
# renaming the column to paid so it makes more sense to whoever is working on the dataset

clean_df.rename(columns={'Type': 'Paid'}, inplace=True)
```

#### Genres Column:

we will be parsing the genres and splitting concatenated genres then turning them into dummies (for the sake of one hot encoding)

```python
# create a new df from our clean df that we are working with

splited = clean_df
```

```python
# spliting the concatenated genres:

splited[['genre_1', 'genre_2']] = clean_df['Genres'].str.split(';', n=1, expand=True)
```

```python
# seeing how many unique genres we have:

split_genres = clean_df['Genres'].str.split(';')

s = set()
for i in split_genres:
    for genre in i:
        s.add(genre)
print(len(s))
```

```python
# getting the dummies from the genres

genres_1 = pd.get_dummies(splited['genre_1'])
genres_2 = pd.get_dummies(splited['genre_2'])
```

```python
# checking if the above step worked:

genres_1
```

```python
genres_2
```

```python
# dropping the genres columns in the dataframe we made so we can instead use the dummies:

splited = splited.drop(['Genres', 'genre_1', 'genre_2'], axis=1)
```

```python
# adding in the dummies to the dataframe:

merged_splits = pd.concat([splited,genres_1, genres_2], axis=1)
```

```python
# checking if the above step worked:

merged_splits.head()
```

#### Turning the Categories into Columns:

```python
# getting the dummies from the category column

categories = pd.get_dummies(clean_df['Category'])
```

```python
# adding the dummies to the df that we will be using as full one hot encoded dataset:

merged_splits = pd.concat([merged_splits, categories], axis=1)
```

```python
# dropping the category column since we have the dummies now:

merged_splits = merged_splits.drop('Category', axis=1)
```

```python
# checking that the above steps worked:

merged_splits.head()
```

#### Content Rating:

```python
# getting the dummies from the content rating column

content_rating = pd.get_dummies(clean_df['Content Rating'])
```

```python
# dropping the content rating column from the one hot data set:

merged_splits = merged_splits.drop('Content Rating', axis=1)
```

```python
# adding the dummies to the dataset:

merged_splits = pd.concat([merged_splits, content_rating], axis=1)
```

```python
# checking if the above steps worked:

merged_splits.head()
```
